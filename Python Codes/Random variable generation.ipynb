##### 1.1 Generating a Gaussian sample using the uniform distirbution 

```python
import numpy as np

# 1. Generating a uniform sample in the interval [0,1]
n = 500 # Sample size
np.random.seed(0) # Fixing the seed for reproducibility
S = np.random.uniform(low = 0, high = 1, size = n)    
```

```python
# 2. Implementing an approximation of the inverse CDF of a standard normal
def inv_CDF_normal(x):
    """
    This function returns the inverse CDF of a standard normal.

    Parameters
    ----------
    x : float
        A value between 0 and 1.

    Returns
    ----------
    float
    """
    # The inverse CDF is defined on the interval (0,1)
    if x <= 0 or x >= 1:
        raise ValueError("x must be a value between 0 and 1")
    
    # Approximation parameters
    a_0, a_1, b_1, b_2 = 2.30753, 0.27061, 0.99229, 0.04481
    t = np.sqrt( -2*np.log(x) )

    F_inv = t - (a_0 + a_1*t)/(1 + b_1*t + b_2*t**2)
    return(F_inv)
```

```python
# 3. Plotting the histogram of the image of the sample with the inverse CDF
import matplotlib.pyplot as plt
import seaborn as sns

S2 = np.array([inv_CDF_normal(s) for s in S])

# Plotting
plt.hist(S2, bins=20, density=True, alpha=0.6, color='g')
sns.kdeplot(S2, fill=True, color='blue')

plt.title('Density Plot of the Image of the Sample S')
plt.xlabel('Value')
plt.ylabel('Density')

plt.show()
```
